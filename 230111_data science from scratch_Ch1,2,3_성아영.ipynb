# 정규표현식

```python
def solution(s):
    result = []

    for i in s.split(" "):
        number = []
        # 전화번호 길이가 맞고 앞에 010이 숫자인지 확인
        if len(i) == 13 and i[:2].isdigit():
            i = '010' + '-' + '****-****'
        number.append(i)
        result.append(" ".join(number))
    return ' '.join(s for s in result)
```

```python
s= '김철수님의 전화번호는 010-1234-5678 이며 이영희님은 010-1001-2222 입니다.'

print(solution(s))
```

```python
import math

class Circle:
    #초기화
    def __init__(self, radius):
        if radius < 0:
            raise ValueError("positive radius expected")
        self.radius = radius
    # 원의 면적 계산
    def area(self):
        # radius가 0 이상일때만 실행하도록
        assert self.radius >= 0, "positive radius expected" 
        return math.pi * self.radius ** 2
    def correct_radius(self, correction_coefficient):
        # radius에 특정 숫자를 곱
        self.radius *= correction_coefficient
```

```python
tire = Circle(30)
tire.area()
```

```python
tire.correct_radius(-1.02)
tire.radius
tire.area()
```

# 종합 코드 예시

```python
def solution(s):
    answer = []
    tmp = s.replace('{', '')
    tmp = tmp.split('}')
    tmp = ' '.join(tmp).split()
    new = []
    
    for i in tmp:
        k = i
        if i[0] == ',':
            k = i[1:]
        new.append(k)
    # 리스트를 길이순으로 정렬
    new = sorted(new, key=len)
    for i in new:
        # ,를 기준으로 나눈 후
        for n in i.split(','):
            # n이 answer에 없으면 n을 추가
            if int(n) not in answer:
                answer.append(int(n))
    
    return answer
```

```python
s = "{{1,2,3},{2,1},{1,2,4,3},{2}}"
solution(s)
```
